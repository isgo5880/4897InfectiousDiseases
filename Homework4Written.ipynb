### Part a

To estimate $R_0$ by examining the period of exponential growth using method 1 from week 9, we need to identify a period of exponential growth, estimate the growth rate, and use that to estimate $R_0$. In early stages of an epidemic, we expect exponential growth rate, which we can see in the first 16 weeks of data:

If we plot this on a log-y axis, which would make the exponential shape a linear shape, we can perform linear regression to estimate the slope of the line. Doing this results in the following, where the line of best fit is described by the equation $y = 0.048 + 0.431x$.

Using this estimated slope, we can then estimate $R_0$ using the equation $$R_0 \approx 1 + \frac{\hat{m}}{\gamma + \mu}$$, which ends up being 1.845. The calculated 95% confidence interval for the slope is [0.3896, 0.4723], which then makes our $R_0$ estimate have a confidence interval of [1.7639, 1.9261].

### Part b

To estimate $R_0$ using seroprevalence data, we can use the seropevalence given to us and calculate $R_0$ like so:
$$R_0 = \frac{1}{1 - seroprevalence}$$

Plugging in the $517/1000$ given from the seroprevalence study, we can estimate $R_0$ to be 2.07 plus or minus 0.000979.

### Part c

To estimate $R_0$ a third way, we can estimate it from equilibrium prevalence. To do this, I chose a window of the weekly new cases that seemed to be approaching/centered around some equilibrium, and then took the mean of the weekly new cases from this window. From there, we know that about 10% of cases are being observed, so to estimate the true prevalence we divide this mean by 10%, and multiply by the duration of the infection, or 2 weeks. We can divide by the population size to get a proportion. The estimated $i_eq$ is about 0.009532. 

Calculating $R_0$ and the 95% confidence interval for this estimate results in an approximate $R_0$ of 1.946 and a confidence interval of [1.8932, 2.0019].

### Part d

All of the $R_0$ estimates are pretty close to one another, between 1.85 and 2.1. The first estimate has a pretty large uncertainty, followed by the third, and the seroprevalence estimate has the smallest confidence interval. All of these estimates are different because we are using different types of data to calculate them. The first estimate is using only the first few weeks of data because we are only capturing the initial exponential growth, whereas the third estimate is using the last few weeks of data because we are trying to capture the incidence rates hitting an equilibrium, all of which introduces varying levels of uncertainty. Also, seroprevalence indicates the percentage of a population that has been exposed, and assumes that we are at an equilibrium of susceptibles.
