# Problem 1

The three plots show a logistic curve, where the first stepsize shows many oscillations that occur arround the analytical solution, and the oscillations decrease in amplitude as the stepsize decreases. The smaller the stepsize, the better approximation we get to the analytical solution.

This plot shows that as stepsize increases, the maximum absolute error also increases. I would want a larger stepsize when I want to converge to an equilibrium faster, but I would want a smaller stepsize to be more accurate in determining what that equilibrium is. In an advisory position I should be aware that with a larger stepsize there is a larger error in my approximation, and my advice should be limited by my error.
